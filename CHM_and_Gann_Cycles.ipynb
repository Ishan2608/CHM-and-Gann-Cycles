# Fetch Historical Data for Each Exchange

```python
def cache_data(func):
  """
  Decorator function that caches the result of the decorated function
  in a global variable 'combined_data'.

  Args:
      func: The function to be decorated.

  Returns:
      The result of the decorated function.
  """
  def wrapper(*args, **kwargs):
    global combined_data, data_exists
    if combined_data is None:
      # Check for file existence (optional)
      try:
        combined_data = pd.read_csv(combined_data_path)
        print("Loaded data from existing CSV file.")
        combined_data_exists = True
      except FileNotFoundError:
        print("CSV file not found. Fetching data...")
        combined_data = func(*args, **kwargs)
        combined_data.to_csv(combined_data_path, index=False)  # Save for future use
    return combined_data
  return wrapper

# Define a function to fetch historical data
@cache_data
def fetch_stock_data(exchange, symbols, start_date, end_date):
    data_frames = []
    for symbol in symbols:
        try:
            # Fetch historical data
            stock_data = yf.download(symbol + exchange, start=start_date, end=end_date)
            # Fill missing values with zero
            stock_data = stock_data[['Close']].fillna(0)
            stock_data.columns = [symbol]
            data_frames.append(stock_data)
        except Exception as e:
            print(f"Error fetching data for {symbol}: {e}")

    # Combine all dataframes
    combined_data = pd.concat(data_frames, axis=1)
    combined_data.reset_index(inplace=True)
    combined_data.rename(columns={'Date': 'DATE'}, inplace=True)

    return combined_data
```

# Implement CHM and Gann

```python
def analyze_bullish_stocks(df, start_date, end_date, thresh):
    # Step 1: Add a 'Year' column to the dataframe
    df['Year'] = pd.to_datetime(df['DATE'], format='%d-%m-%Y').dt.year

    # Step 2: Filter the data for the specified time period across all years
    def filter_by_period(df, start_date, end_date):
        df['Month-Day'] = pd.to_datetime(df['DATE'], format='%d-%m-%Y').dt.strftime('%m-%d')
        return df[(df['Month-Day'] >= start_date) & (df['Month-Day'] <= end_date)]

    filtered_data = filter_by_period(df, start_date, end_date)

    # Step 3: Group by Year and perform analysis
    bullish_stocks = {}
    for stock in df.columns[1:-1]:  # Skipping 'DATE' and 'Year' columns
        price_changes = []

        # Analyze each year's price change during the specified period
        for year, group in filtered_data.groupby('Year'):
            start_price = group.iloc[0][stock]
            end_price = group.iloc[-1][stock]

            if start_price > 0:  # Avoid division by zero or non-existent data
                change = (end_price - start_price) / start_price * 100  # Percentage change
                price_changes.append(change)

        # Step 4: Determine if the stock is bullish (>70% increase)
        if price_changes and sum(1 for change in price_changes if change > 0) / len(price_changes) > 0.7:
            avg_increase = sum(price_changes) / len(price_changes)  # Average percentage increase
            bullish_stocks[stock] = avg_increase

    # Step 5: Output the list of bullish stocks and their average increase
    bullish_df = pd.DataFrame(bullish_stocks.items(), columns=['Stock', 'Avg_Increase'])
    filtered_df = bullish_df[bullish_df['Avg_Increase'] >= thresh]
    print("Bullish stocks with their average increase are listed.")
    return filtered_df
```

# Implement CHM on Specific Stocks

```python
def analyze_specific_stocks(combined_data, start_date, end_date, stock_list, threshold=6):
    # Step 1: Ensure the 'DATE' column is in datetime format
    combined_data['DATE'] = pd.to_datetime(combined_data['DATE'], errors='coerce')
    
    # Step 2: Add a 'Year' column to the dataframe for yearly grouping
    combined_data['Year'] = combined_data['DATE'].dt.year

    # Step 3: Filter the data for the specified time period across all years
    def filter_by_period(df, start_date, end_date):
        # Extract month and day from the 'DATE' column
        df['Month-Day'] = df['DATE'].dt.strftime('%m-%d')
        return df[(df['Month-Day'] >= start_date) & (df['Month-Day'] <= end_date)]

    # Apply the filter
    filtered_data = filter_by_period(combined_data, start_date, end_date)

    # Step 4: Group by Year and perform analysis
    bullish_stocks = {}
    for stock in stock_list:
        price_changes = []

        # Analyze each year's price change during the specified period
        for year, group in filtered_data.groupby('Year'):
            if stock not in group.columns:
                continue  # Skip if stock column is not present

            # Ensure there are valid (non-NaN) prices for the stock
            start_price = group.iloc[0][stock]
            end_price = group.iloc[-1][stock]
            
            if pd.isna(start_price) or pd.isna(end_price):
                continue  # Skip if either price is missing
            
            if start_price > 0:  # Avoid division by zero or non-existent data
                change = (end_price - start_price) / start_price * 100  # Percentage change
                price_changes.append(change)

        # Step 5: Determine if the stock is bullish (increase in >70% of the periods)
        if price_changes and sum(1 for change in price_changes if change > 0) / len(price_changes) > 0.7:
            avg_increase = sum(price_changes) / len(price_changes)  # Average percentage increase
            bullish_stocks[stock] = avg_increase

    # Step 6: Create a DataFrame of bullish stocks and their average increase
    bullish_df = pd.DataFrame(bullish_stocks.items(), columns=['Stock', 'Avg_Increase'])

    # Step 7: Filter based on the user-provided threshold
    filtered_df = bullish_df[bullish_df['Avg_Increase'] >= threshold]
    
    # Print the output for debugging (optional)
    print(f"Bullish stocks with Avg_Increase >= {threshold}%:")
    print(filtered_df)

    return filtered_df
```

# Update Data

```python
def update_stock_data(combined_data, symbols, exchange):
    """
    Updates the stock data in the combined_data dataframe with the latest data.

    Args:
        combined_data: The existing combined data dataframe.
        symbols: A list of stock symbols.
        exchange: The exchange code (e.g., '.NS' for NSE, '.BO' for BSE).

    Returns:
        The updated combined data dataframe.
    """

    # Find the last date in the combined data
    last_date = combined_data['DATE'].max()

    # Get the current date
    current_date = datetime.now().strftime('%Y-%m-%d')

    # Fetch the new data from the last date to the current date
    new_data = fetch_stock_data(exchange, symbols, last_date, current_date)

    # Merge the new data with the existing data
    combined_data = pd.merge(combined_data, new_data, on='DATE', how='outer')

    # Remove duplicate symbols by keeping only the first occurrence
    combined_data = combined_data.loc[:,~combined_data.columns.duplicated()]

    return combined_data

# Example usage:
# Assuming combined_data is your existing dataframe
combined_data = update_stock_data(combined_data, NSE_SYMBOLS, '.NS')
combined_data = update_stock_data(combined_data, BSE_SYMBOLS, '.BO')

# Save the updated data to CSV and Excel
combined_data.to_csv('combined_stock_data.csv', index=False)
combined_data.to_excel('combined_stock_data.xlsx', index=False)
```
